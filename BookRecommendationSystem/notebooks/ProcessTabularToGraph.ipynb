```python
import pandas as pd
import torch
import torch_geometric as torch_g
```

```python
df = pd.read_csv("../data/processed/BookReviewProcessedData.csv")
```

```python
summary_attr = [f"Summary_{i}" for i in range(1, 35)]
```

```python
user_attributes = ["user_id", "age", "city", "state", "country"]
book_attributes = ["isbn", "book_title", "book_author", "year_of_publication"] + summary_attr
```

```python
users_df = df[user_attributes].reset_index(drop=True)
books_df = df[book_attributes].reset_index(drop=True)
```

```python
books_df.head()
```

```python
users_df.head()
```

```python
rating = df[["user_id", "isbn", "rating"]].reset_index(drop=True)
```

```python
rating
```

```python
y = rating["rating"].to_numpy()
```

```python
data = torch_g.data.HeteroData()
```

```python
edge_index = rating[["user_id", "isbn"]].values.transpose()
```

```python
data["users"].x = users_df.to_numpy()
data["books"].x = books_df.to_numpy()
```

```python
data["users", "rating", "books"].edge_index = edge_index
data["users", "books"].y = y
```
